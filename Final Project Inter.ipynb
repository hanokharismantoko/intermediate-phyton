# 5.Get rid of the missing value:


## a. Drop columns which has missing value that cannot be filled with any logical value (like name, host_id, etc)


```python
data["last_review"] = pd.to_datetime(data["last_review"],format='%Y-%m-%d %H:%M', errors = 'coerce')
data.dtypes
```

```python
data
```

```python
data.dropna(axis=0, how='all', inplace=True)
data
```

## b. Fill the missing value in last_review with the most common date


```python
data['last_review'] = data['last_review'].fillna(data['last_review'].mode()[0])
data
```

## c. Fill the missing value in reviews_per_month with the mean


```python
data['reviews_per_month'] = data['reviews_per_month'].fillna(data['reviews_per_month'].mean())
data
```

# 8. Visualize the price to compare the price of available rooms


```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
price_x = data["price"]
room_y = data["availability_365"]
#width = 1
#pv_room_y = data[]
```

```python
plt.bar(room_y, price_x, color='salmon')
plt.title('price vs available rooms')
plt.xlabel('available rooms')
plt.ylabel('price')
plt.legend()
plt.show()
```

```python
plt.bar(price_x, room_y, color='salmon')
plt.title('price vs available rooms')
plt.xlabel('price')
plt.ylabel('availble rooms')
plt.legend()
plt.show()
```

```python
plt.scatter(room_y, price_x, s=50, c=v, cmap='pink', edgecolor='black',
            linewidth=1, alpha=0.75)

cbar= plt.colorbar()
cbar.set_label('availability')

plt.title('price vs available rooms')
plt.xlabel('price')
plt.xlabel('available rooms')

plt.tight_layout()
plt.show()
```
